```python
# Day 1: PE 估值判断
pe_ratio = 25
if pe_ratio <= 0:
    print(f"PE = {pe_ratio}，估值无效")
elif pe_ratio < 10:
    print(f"PE = {pe_ratio}，估值偏低")
elif pe_ratio < 25:
    print(f"PE = {pe_ratio}，估值合理")
elif pe_ratio < 40:
    print(f"PE = {pe_ratio}，估值偏高")
else:
    print(f"PE = {pe_ratio}，估值过高")
```
